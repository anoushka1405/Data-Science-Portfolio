```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate dates & values
dates = pd.date_range(start='1/1/2023', periods=10, freq='D')
values = np.random.randn(10).cumsum()

# Create plot
plt.figure(figsize=(8, 5))  # Adjust figure size
plt.plot(dates, values, marker='o', linestyle='-', color='green', label="Data")

# Add trendline 
z = np.polyfit(range(len(values)), values, 1)
p = np.poly1d(z)
plt.plot(dates, p(range(len(values))), linestyle='dashed', color='red', label="Trendline")

# Formatting
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Plot")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.normal(0, 0.1, 50)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Function to plot different graphs
def plot_graph(plot_type):
    plt.figure(figsize=(6, 4))

    if plot_type == "line":
        plt.plot(x, y, marker="o", linestyle="-", color="blue", label="Line Plot")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")

    elif plot_type == "scatter":
        plt.scatter(x, y, color="red", alpha=0.6, label="Scatter Plot")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")

    elif plot_type == "bar":
        plt.bar(categories, values, color="green", label="Bar Plot")
        plt.xlabel("Categories")
        plt.ylabel("Values")

    elif plot_type == "pie":
        plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=90, colors=["red", "blue", "green", "orange"])
    
    elif plot_type == "histogram":
        data = np.random.randn(1000)
        plt.hist(data, bins=20, color="purple", alpha=0.7, edgecolor="black", label="Histogram")
        plt.xlabel("Values")
        plt.ylabel("Frequency")

    elif plot_type == "box":
        data = [np.random.randn(100) for _ in range(4)]
        plt.boxplot(data, labels=categories, patch_artist=True, notch=True)
    
    elif plot_type == "bubble":
        sizes = np.random.randint(20, 300, 50)
        plt.scatter(x, y, s=sizes, color="cyan", alpha=0.5, edgecolors="black", label="Bubble Plot")

    plt.title(f"{plot_type.capitalize()} Plot")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()

# List of plot types
plot_types = ["line", "scatter", "bar", "pie", "histogram", "box", "bubble"]

# Loop through and generate each plot
for plot in plot_types:
    plot_graph(plot)
```

```python
def plot_function(x, y, plot_type="line", color="blue"):
    plt.figure(figsize=(6, 4))
    
    if plot_type == "line":
        plt.plot(x, y, color=color, label="Line Plot")
    elif plot_type == "scatter":
        plt.scatter(x, y, color=color, alpha=0.6, label="Scatter Plot")
    
    plt.legend()
    plt.title(f"{plot_type.capitalize()} Plot")
    plt.show()

# Generate data
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Call function
plot_function(x, y, plot_type="line")
plot_function(x, y, plot_type="scatter", color="red")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_all_graphs():
    # Creating the DataFrame
    df = pd.DataFrame({
        'A': np.random.randn(10),
        'B': np.random.randn(10)
    })

    # Line Plot
    df.plot(kind='line', title="Pandas Line Plot")
    plt.show()

    # Bar Chart
    df.plot(kind='bar')
    plt.title("Bar Chart")
    plt.show()

    # Histogram
    df.plot(kind='hist')
    plt.title("Histogram")
    plt.show()

    # Box Plot
    df.plot(kind='box')
    plt.title("Box Plot")
    plt.show()

    # Scatter Plot
    df.plot(kind='scatter', x='A', y='B')
    plt.title("Scatter Plot")
    plt.show()

    # Pie Chart
    df['A'] = abs(df['A'])  # Ensure non-negative values
    df['A'].plot(kind='pie', autopct='%1.2f%%')  # 2 decimal places
    plt.title("Pie Chart")
    plt.ylabel("")  # Remove default label
    plt.show()

# Call the function
plot_all_graphs()
```

```python
def graph(kind):
    if kind == 'line':
        x = np.linspace(1,10,100)
        y = np.random.randn(100)
        plt.plot(x,y)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title("Line Plot")
        plt.show()
    elif kind == 'bar':
        x = np.linspace(1,10,100)
        y = np.random.randn(100)
        plt.bar([x,y],color = 'green')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title("Bar Plot")
        plt.show()
    elif kind == 'pie':
        x = ['50','25','25']
        y = ['A','B','C']
        explode = [0,0.1,0]
        colors = ['red','blue','green']
        plt.pie(x,labels = y, colors = colors, explode = explode, autopct = '%1.1f%%', shadow = True, startangle = 90)
        plt.title("Pie Plot")
        plt.show()
    elif kind == 'box':
        x = np.random.randn(100)
        y = np.random.randn(100)
        plt.boxplot([x, y])
        plt.title("Box Plot")
        plt.xticks([1, 2], ['x', 'y'])  # Optional: labeling the boxes
        plt.show()
    elif kind=='histogram':
        x = np.linspace(1,10,100)
        y = np.random.randn(100)
        plt.hist(y,bins = 30, color = 'purple')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title("Histogram")
        plt.show()
    elif kind == 'time-series':
        a = pd.date_range(start = '1/1/2023', periods = 10, freq = 'D')
        b = np.random.randn(10).cumsum()
        plt.plot(a,b)
        plt.title("Time Series Plot")
        plt.show()
    elif kind == 'area':
        x = np.linspace(1,10,100)
        plt.fill_between(x,np.sin(x), color = 'powderblue', alpha = 0.5)
        plt.plot(np.linspace(1,10,100), np.sin(x), color = 'slateblue', alpha = 0.7)
        plt.title("Area Plot")
        plt.show()
    elif kind == 'scatter':
        x = np.random.randint(1,100,size=50)
        y = np.random.randint(1,100,size=50)
        plt.scatter(x,y,color='red')

print('List of graphs : [ line, bar, pie, box, histogram, time-series, area, scatter ]')
kind = input("Enter type of graph")
graph(kind)

                 
        
        


        
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'name': ['Anoushka', 'Rahul', 'Meera', 'Kabir'],
    'total_marks': [92, 75, 88, 60]
}

df = pd.DataFrame(data)

# Highlight function
def highlight_marks(s):
    '''
    Highlights marks greater than 80 in red and others in green.
    '''
    return ['background-color: red' if v > 80 else 'background-color: green' for v in s]

# Apply style
styled_df = df.style.apply(highlight_marks, subset=['total_marks'])

# Display in Jupyter or web-based interface
styled_df
```
