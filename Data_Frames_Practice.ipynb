```python
import pandas as pd

# Step 1: Create the original wide-format DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Math': [85, 90],
    'Science': [95, 80]
})

print("Original DataFrame (wide format):")
print(df)
print("\n")

# Step 2: Use melt() to convert from wide to long format
melted = pd.melt(df, id_vars=['Name'], var_name='Subject', value_name='Score')
print("DataFrame after melt() (long format):")
print(melted)
print("\n")

# Step 3: Use pivot() to convert back from long to wide format
pivoted = melted.pivot(index='Name', columns='Subject', values='Score')
print("DataFrame after pivot() (wide format again):")
print(pivoted)
print("\n")

# Step 4: Optional - Reset index to make 'Name' a regular column again
final_df = pivoted.reset_index()
print("Final DataFrame with reset index:")
print(final_df)

```

```python
import numpy as np

# Given data
students = [
    ('Anoushka', 162.5, '10A'),
    ('Ravi', 170.2, '10B'),
    ('Meera', 158.4, '10A'),
    ('Karan', 175.0, '10C')
]

# Given data types (name: string, height: float, class: string)
dtype = [('name', 'U20'), ('height', 'f4'), ('class', 'U10')]

# Create structured array
student_array = np.array(students, dtype=dtype)

# Sort based on height
sorted_students = np.sort(student_array, order='height')

# Print the result
print("Students sorted by height:")
print(sorted_students)
```

```python
import numpy as np

# Create a 3x3 identity matrix (unitary in real numbers)
A = np.identity(3)

# Vertical stack: two matrices one below the other (6x3)
vstacked = np.vstack([A, A])

# Horizontal stack: two matrices side by side (3x6)
hstacked = np.hstack([A, A])

# Print results
print("Original 3x3 Unitary Matrix:\n", A)
print("\nVertically Stacked (6x3):\n", vstacked)
print("\nHorizontally Stacked (3x6):\n", hstacked)
```

```python
import pandas as pd
import numpy as np

def create_pivot_table(df):
    """
    Takes a DataFrame with 'date' and 'revenue' columns,
    calculates monthly revenue, creates a pivot table by year and month,
    and adds a 'Mean' row.

    Args:
        df (pd.DataFrame): Input DataFrame with 'date' and 'revenue' columns.

    Returns:
        pd.DataFrame: Pivot table with years as index, months as columns,
                      revenue as values, and a 'Mean' row.
    """
    df['date'] = pd.to_datetime(df['date'])
    df['Year'] = df['date'].dt.year
    df['Month'] = df['date'].dt.strftime('%b')

    pivot_df = df.pivot_table(index='Year', columns='Month', values='revenue', aggfunc='sum', fill_value=0)
    pivot_df = pivot_df.reindex(columns=['Jan', 'Apr', 'Aug', 'Sep', 'Oct'], fill_value=0)
    pivot_df.loc['Mean'] = pivot_df.mean(axis=0)
    pivot_df = pivot_df.round(2)
    pivot_df.reset_index(inplace=True)
    return pivot_df

# Sample DataFrame (replace with your actual df)
data = {'order_id': ['id01', 'id02', 'id03', 'id04', 'id05', 'id06', 'id07', 'id08', 'id09', 'id10'],
        'date': ['2018-01-01', '2019-01-07', '2019-10-21', '2019-04-04', '2019-08-03',
                 '2020-06-28', '2020-09-08', '2019-09-04', '2018-10-20', '2020-09-29'],
        'revenue': [62.98, 51.83, 31.69, 16.38, 492, 169, 213, 263, 139, 364]}
df = pd.DataFrame(data)

output_pivot_table = create_pivot_table(df.copy()) # Use a copy to avoid modifying the original df
print(output_pivot_table)
```
